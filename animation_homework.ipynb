```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
rc('animation', html='html5')
```

```
fig = plt.figure()
ax = plt.axes(xlim=(-2.0, 2.0), ylim=(-1.6, 1.6))

ax.set_xticks([-2.0,-1.0, 0, 1.0, 2.0])
ax.set_xticklabels(['$-2$','$-1$','$0$','$1$','$2$'])
ax.set_yticks([-1.5, -1,-0.5,0,0.5, 1, 1.5])
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.close()
```

```
line, = ax.plot([], [], lw=2)
def init():
    line.set_data([], [])
    return (line,)

def animate(t):
    x = np.linspace(-2.0, 2.0, 300)
    y = np.sin(2*np.pi*x)*np.exp(-0.1*t)
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init, interval=30, blit=True)
anim
```
