```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
rc('animation', html='html5')
```

```
fig = plt.figure()
ax = plt.axes(xlim=(-6.0*np.pi, 6.0*np.pi), ylim=(-1.6, 1.6))

ax.set_xticks([-6.0*np.pi,-3.0*np.pi, 0, 3.0*np.pi, 6.0*np.pi])
ax.set_xticklabels(['$-6\pi$','$-3\pi$','$0$','$3\pi$','$6\pi$'])
ax.set_yticks([-1.5, -1,-0.5,0,0.5, 1, 1.5])
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.close()
```

```
line, = ax.plot([], [], lw=2)
def init():
    line.set_data([], [])
    return (line,)

def animate(t):
    x = np.linspace(-6.0*np.pi, 6.0*np.pi, 300)
    y = np.sin(2*np.pi*x)*np.exp(-0.1*t)
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init, interval=30, blit=True)
anim
```
